---

## Setup

```raku
use Graph::RandomMaze;
```

```raku
sink my $engine = 'neato';
```

---

## Rectangular maze

```raku
my %res = random-maze(:10rows, :20columns);

deduce-type(%res)
```

```raku
#%html
%res<walls>.dot(:$engine, :8size, vertex-shape => 'point', edge-thickness => 12):svg
```

```raku
#%html
my $gSolution = %res<paths>.subgraph(%res<solution>);
my $g = %res<walls>.union($gSolution);
my $highlight = $gSolution;
$g.dot(
    :$highlight,
    :$engine, 
    :8size, 
    vertex-shape => 'point', 
    edge-thickness => 12
):svg
```
